# Eigenvalue/Eigenvector Calculation using Sparse Matrix

```python
print(Titanic_dataset1.isna().sum())
```

```python
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix
from sklearn.decomposition import TruncatedSVD
```

```python
# Select relevant columns
Titanic_dataset1 = Titanic_dataset1[['Survived', 'Age', 'Pclass', 'Fare', 'Age_group_num', 'Sex_male']]
```

```python
# Convert categorical variables to numerical
Titanic_dataset1['Age_group_num'] = pd.Categorical(Titanic_dataset1['Age_group_num']).codes
Titanic_dataset1['Sex_male'] = pd.Categorical(Titanic_dataset1['Sex_male']).codes
```

```python
# Drop rows with missing values
Titanic_dataset_dropped = Titanic_dataset1.dropna()
```

```python
# Convert dataset to sparse matrix
sparse_matrix = csr_matrix(Titanic_dataset_dropped)
```

```python
# Use TruncatedSVD for sparse data
svd = TruncatedSVD(n_components=5)
svd.fit(sparse_matrix)
```

```python
eigenvalues = svd.explained_variance_
eigenvectors = svd.components_
print(f'Eigenvalues: {eigenvalues}')
print(f'Eigenvectors: {eigenvectors}')
```

# Creating a Custom Aggregation Function

```python
# Select only numeric columns
numeric_columns = Titanic_dataset1.select_dtypes(include=[np.number])
Titanic_dataset1
```

```python
# Define custom aggregation functions
def survival_rate(x):
    """Calculate survival rate"""
    return (x.sum() / len(x)) * 100

def fare_avg(x):
    """Calculate average fare"""
    return x.mean()

def age_std(x):
    """Calculate standard deviation of age"""
    return x.std()

# Apply custom aggregation functions using groupby
grouped = Titanic_dataset1.groupby('Pclass').agg({
    'Survived': survival_rate,
    'Fare': fare_avg,
    'Age': age_std
})

print(grouped)
```

# Multiple Y-Axes Sharing a Common X-Axis Plo

```python
import matplotlib.pyplot as plt

# Assuming 'grouped' DataFrame is already created

# Create a figure and axis
fig, ax1 = plt.subplots()

# Plot Survived on the left axis
ax1.plot(grouped.index, grouped['Survived'], color='blue')
ax1.set_xlabel('Age Group')
ax1.set_ylabel('Survival Rate', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
# Show legend
ax1.legend(['Survived'], loc='upper left')

# Create a second axis on the right
ax2 = ax1.twinx()

# Plot Fare on the right axis
ax2.plot(grouped.index, grouped['Fare'], color='red')
ax2.set_ylabel('Average Fare', color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Set title
plt.title('Survival Rate and Average Fare by Age Group')

# Combine legends from both axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper center')

# Show plot
plt.show()

```
